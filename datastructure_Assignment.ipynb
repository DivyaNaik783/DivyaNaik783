Q13.How do sets handle duplicate values in Python?
 Ans:Hashing Mechanism: Sets in Python are implemented using hash tables, and each element must be hashable. The hash value of an element determines its position in the set. When a new element is added, the set checks if the element already exists by comparing hash values. If the hash value is the same, the element is considered a duplicate and is not added again.

Efficient Membership Testing: Sets are optimized for fast membership testing (O(1) on average), and maintaining uniqueness simplifies this process. Duplicate entries are unnecessary and could complicate the set's operations.


Q14.How does the “in” keyword work differently for lists and dictionaries?

 Ans:The in keyword in Python is used to check for membership, but its behavior differs when used with lists versus dictionaries:
 For Lists:The in keyword checks whether a value exists in the list and
  It performs a sequential search through the list, so its performance is O(n) in the worst case
  For Dictionaries:The in keyword checks whether a key exists in the dictionary, not the values and It performs a hash lookup, so its performance is O(1) on average.

Q19.Why are dictionaries considered unordered, and how does that affect data retrieval?

 Ans:Dictionaries in Python are considered unordered because they do not store their elements in a fixed, predictable sequence based on insertion or any inherent order of the keys. Instead, dictionaries organize their data using a hash table, which optimizes operations like lookups, insertions, and deletions. This impacts how data is stored and retrieved.

Q20.Explain the difference between a list and a dictionary in terms of data retrieval.


 For lists:Data Access By index (positional access).
 For Dictionary:Data Access By key (key-value mapping).

 For lists:Access Speed is O(1) for indexing.
 For Dictionary: Access speed is O(1) average for key lookup, O(n) worst-case due to hash collisions.

 For lists:Structure is Ordered collection of values.
 For Dictionary:Structure is Key-value pairs, with keys being unique.

 For lists:Efficiency is Fast sequential access; good for ordered data.
 For Dictionary:Efficiency is Optimized for fast lookups and updates via keys.

 For lists:Suitable for homogeneous or heterogeneous sequences.
 For Dictionary:Better for representing relationships or mappings.
